```python
import pandas as pd

# Load the dataset
df = pd.read_csv('~/Sales-Analysis-Project/data/superstore_cleaned.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn
sns.set(style="whitegrid")
# Set the figure size
plt.figure(figsize=(12, 6))
# Create a bar plot
sns.barplot(x='Year', y='Sales', data=df, hue='Month', palette='viridis')
# Set the title and labels
plt.title('Sales by Year and Month')
plt.xlabel('Year')
plt.ylabel('Sales')
# Show the plot
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Set the figure size for the second plot
plt.figure(figsize=(12, 6))
# Create a bar plot for regions
sns.barplot(x='Region', y='Sales', data=df, palette='viridis', hue='Segment')
# Set the title and labels
plt.title('Sales by Region and Segment')
plt.xlabel('Region')
plt.ylabel('Sales')
# Show the plot
plt.legend(title='Segment', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Set the figure size for the third plot
plt.figure(figsize=(12, 6))
# Create a bar plot for categories
sns.barplot(x='Sub-Category', y='Sales', data=df, palette='viridis', hue='Category')
# Set the title and labels
plt.title('Sales by Category and Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
# Rotate the x labels for better readability
plt.xticks(rotation=45)
# Show the plot
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
